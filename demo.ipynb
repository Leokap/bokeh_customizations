```python
from bokeh.plotting import show, save
from bokeh.io import output_notebook


from bokeh_customizations import standard_figure, insert_line_plot_with_segmented_coloring, apply_click_policy_mute_others
import numpy as np


output_notebook(hide_banner=True)

# Create some dummy data
x = np.linspace(0, 10, 100)
y = np.sin(x)
color_indices = (x // 2).astype(int)

# Create a standard figure
p = standard_figure(title="New Bokeh Customizations Demo")

# Insert a segmented line plot
insert_line_plot_with_segmented_coloring(p, x, y, color_indices, legend_labels=np.arange(6).astype(str))#, legend_label="Sin Wave")

# Add two other normal line to test "mute others"
p.line(x, np.cos(x), line_color="black", line_dash="dashed", legend_label="Cos Wave")
p.line(x, np.cos(x)/2, line_color="green", line_dash="dashed", legend_label="Cos Wave halfed")

# Apply mute others policy
apply_click_policy_mute_others(p)

# In a real environment, you would call show(p)
show(p)
print("Plot created successfully with KeyboardShortcutsTool.")
```

```python
from bokeh.plotting import show, output_file
from bokeh_customizations.plotting import get_column_layout_with_shared_x
import numpy as np
import pandas as pd

# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Create a column layout with 3 plots sharing the same x-axis
layout = get_column_layout_with_shared_x(
    n=3,
    titles=["Sine Wave", "Cosine Wave", "Sine × Cosine"],
    width=800,
    height=200
)

# Add data to each plot
layout.children[0].line(x, y1, line_width=2, color="blue")
layout.children[1].line(x, y2, line_width=2, color="red")
layout.children[2].line(x, y3, line_width=2, color="green")

# Output and show
show(layout)
```
